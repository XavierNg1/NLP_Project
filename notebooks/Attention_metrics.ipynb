In this notebook we will compute the graphics for the Attention model training

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_json('../logs/model_attention_logs.json').set_index('epoch')
```

```python
df.head(2)
```

```python
plt.figure(figsize=(10, 6))
plt.clf
plt.plot(df.loss.values, '-' )
plt.plot(df.val_loss.values, '-' )
plt.axis('tight')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(('train loss', 'val loss'))
plt.grid()
plt.show()
```
